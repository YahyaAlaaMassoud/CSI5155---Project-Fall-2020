```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import copy

from itertools import product
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.naive_bayes import GaussianNB # Import Naive Bayes Classifier
from sklearn.svm import SVC, LinearSVC # Import SVM Classifier
from sklearn.neighbors import KNeighborsClassifier # Import KNN Classifier
from sklearn.multiclass import OneVsOneClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.preprocessing import label_binarize
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn import tree
from pprint import pprint
from imblearn.over_sampling import SMOTE, ADASYN
from imblearn.under_sampling import ClusterCentroids
from imblearn.combine import SMOTEENN, SMOTETomek
from sklearn.model_selection import KFold, cross_val_score, StratifiedKFold
from prettytable import PrettyTable
from scipy import stats
from matplotlib.pyplot import pie, axis, show
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer, SimpleImputer
from sklearn.feature_selection import VarianceThreshold, SelectKBest, mutual_info_classif, SelectFromModel, RFE
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier
from sklearn.model_selection import cross_validate
from sklearn.metrics import recall_score

random.seed(0)
```

```python
# read the csv file
data = pd.read_csv('dataset_diabetes/diabetic_data.csv', sep=',')
data.head(5)
```

```python
print('Number of records:', data.shape[0], 'Number of features:', data.shape[1])
```

```python
pprint(data.columns)
```

```python
t = PrettyTable(['Feature', 'Number of Unique Elements'])
# list the unique elements in columns
for col in data.columns:
    t.add_row([col, len(data[col].unique())])
print(t)
data.replace(to_replace="?", value=np.nan, inplace=True)
```

```python
data = data[data['diag_1'].notna()]
data = data[data['diag_2'].notna()]
data = data[data['diag_3'].notna()]
data = data[data['race'].notna()]

print('Percentage (%) of Missing values in each feature')
print('--------------------------------------------')
print(np.round(data.isnull().sum() / len(data) * 100, 1))
```

## We will remove the following feature:
### - 'weight' as it has 96.9% missing values, it is highly sparse, and could not be treated directly.
### - 'payer_code', highly sparse as well, with 39.6 missing values, and task irrelevant information
### - 'encounter_id', 'patient_id' redundant and unimportant to the task
## The feature 'medical_speciality' has around 49% missing values, but because it might be relevant to our task, we will use imputation later in the notebook to fill the missing values in this column.

```python
del data['weight'], data['payer_code'], data['encounter_id'], data['patient_nbr']
```

```python
data['age'].value_counts().plot.pie()
```

```python
col_names = data.columns
```

```python
data.diag_1 = data.diag_1.astype(str)
data.diag_2 = data.diag_2.astype(str)
data.diag_3 = data.diag_3.astype(str)
diag_dict = {}
# Circulatory
diag_dict[785] = 'Circulatory'
for i in range(390, 459):
    diag_dict[i] = 'Circulatory'
# Respiratory
diag_dict[786] = 'Respiratory'
for i in range(460, 519):
    diag_dict[i] = 'Respiratory'
# Digestive
diag_dict[787] = 'Digestive'
for i in range(520, 579):
    diag_dict[i] = 'Digestive'
# Injury
for i in range(800, 999):
    diag_dict[i] = 'Injury'
# Musculoskeletal
for i in range(710, 739):
    diag_dict[i] = 'Musculoskeletal'
# Genitourinary
for i in range(580, 629):
    diag_dict[i] = 'Genitourinary'
# Neoplasms
diag_dict[780] = 'Neoplasms'
diag_dict[781] = 'Neoplasms'
diag_dict[782] = 'Neoplasms'
diag_dict[784] = 'Neoplasms'
for i in range(140, 239):
    diag_dict[i] = 'Neoplasms'
for i in range(790, 799):
    diag_dict[i] = 'Neoplasms'
for i in range(240, 279):
    if i is 250:
        continue
    diag_dict[i] = 'Neoplasms'
for i in range(680, 709):
    diag_dict[i] = 'Neoplasms'
# Diabetes
for i in data.diag_1:
    if i.startswith("250"):
        diag_dict[i] = 'Diabetes'
for i in data.diag_1:
    if i not in diag_dict:
        diag_dict[i] = 'Other'

final_dict = {}
for k, v in diag_dict.items():
    final_dict[str(k)] = v

final_dict = pd.Series(final_dict)
data.diag_1 = data.diag_1.map(final_dict)
data.diag_2 = data.diag_2.map(final_dict)
data.diag_3 = data.diag_3.map(final_dict)
# print(len(data.diag_1.unique()))
# print(len(data.diag_2.unique()))
# print(len(data.diag_3.unique()))
```

```python
# label encoding for only columns with categorical data
data.replace(to_replace=np.nan, value='?', inplace=True)
for col in data.columns:
    if col not in ['time_in_hospital', 
                   'num_lab_procedures', 
                   'num_procedures', 
                   'num_medications', 
                   'number_outpatient', 
                   'number_emergency', 
                   'number_inpatient', 
                   'number_diagnoses']:
        unique_elem = sorted(data[col].unique())
        if '?' in unique_elem:
            unique_elem = unique_elem[1:]
        data[col].replace(unique_elem, list(range(len(unique_elem))), inplace=True)
data.replace(to_replace='?', value=np.nan, inplace=True)
```

```python
imputed_data = IterativeImputer(max_iter=100, random_state=0, initial_strategy='most_frequent').fit_transform(data.to_numpy())
# imputed_data = SimpleImputer(strategy='most_frequent').fit_transform(data.to_numpy())
data = pd.DataFrame.from_records(imputed_data)
data.columns = col_names
```

```python
data.head()
```

```python
ids = data.columns[:-1]
data[ids] = (data[ids] - data[ids].mean()) / (data[ids].std() + 1e-4)
data.head(5)
```

```python
features = data.columns[:-1]
X = np.array(data[features])#[:500] # Features
y = np.array(data[data.columns[-1]]).astype(int)#[:500] # Target variable
print("Input data shape", X.shape, "Output data shape", y.shape)
```

```python
kfold = StratifiedKFold(10, random_state=False, shuffle=True)

folds = kfold.split(X, y)
folds = [(train, test) for train, test in folds]
```

```python
estimator = SVR(kernel="linear")
selector = RFE(estimator, n_features_to_select=25, step=1)
selected = selector.fit(X[folds[0][1]], y[folds[0][1]])
print(selected.support_)
print(selected.ranking_)
```

```python
for tak, (i, col) in zip(selected.support_, enumerate(data.columns[:-1])):
    if tak:
        print(col, i)
# gender 1
# age 2
# admission_type_id 3
# discharge_disposition_id 4
# admission_source_id 5
# num_procedures 9
# num_medications 10
# number_outpatient 11
# number_emergency 12
# number_inpatient 13
# diag_3 16
# number_diagnoses 17
# max_glu_serum 18
# A1Cresult 19
# repaglinide 21
# nateglinide 22
# chlorpropamide 23
# glimepiride 24
# glipizide 26
# pioglitazone 29
# acarbose 31
# miglitol 32
# tolazamide 34
# glipizide-metformin 39
# diabetesMed 44
```

```python
X_rfe = selector.transform(X)
X_rfe.shape
```

```python
X_cs, y_cs = SMOTEENN(random_state=0).fit_resample(X_rfe, y)
print(X_cs.shape, y_cs.shape)
```

```python
kfold = StratifiedKFold(10, random_state=False, shuffle=True)

kfolds = kfold.split(X_cs, y_cs)
folds = [(train, test) for train, test in kfolds]
```

```python
# Create Extreme learning Trees Classifier object
xgboost = GradientBoostingClassifier(random_state=0)

xgboost_scores = [xgboost.fit(X_cs[train], y_cs[train]).score(X_cs[test], y_cs[test]) for train, test in folds]
pprint(xgboost_scores)
print('XGBoost', np.mean(xgboost_scores))
```

```python
# Create Extreme learning Trees Classifier object
adaboost = AdaBoostClassifier(random_state=0)

adaboost_scores = [adaboost.fit(X_cs[train], y_cs[train]).score(X_cs[test], y_cs[test]) for train, test in folds]
pprint(adaboost_scores)
print('AdaBoost', np.mean(adaboost_scores))
```

```python
# Create Extreme learning Trees Classifier object
elt_clf = ExtraTreesClassifier()

elt_scores = [elt_clf.fit(X_cs[train], y_cs[train]).score(X_cs[test], y_cs[test]) for train, test in folds]
pprint(elt_scores)
print('XTRA TRees', np.mean(elt_scores))
```

```python
print("XGBoost    Accuracy: %0.2f (+/- %0.2f)" % (np.mean(xgboost_scores), np.std(xgboost_scores) * 2))
print("AdaBoost   Accuracy: %0.2f (+/- %0.2f)" % (np.mean(adaboost_scores), np.std(adaboost_scores) * 2))
print("XTRA Trees Accuracy: %0.2f (+/- %0.2f)" % (np.mean(elt_scores), np.std(elt_scores) * 2))
```

```python
scoring = ['precision_macro', 'recall_macro', 'f1_macro', 'balanced_accuracy', 'accuracy']
scores = cross_validate(estimator=elt_clf,
                         X=X_cs,
                         y=y_cs,
                         cv=folds,
                         scoring=scoring)
print("XTRA Trees Accuracy: %0.2f (+/- %0.2f)" % (np.mean(scores['test_accuracy']), np.std(scores['test_accuracy']) * 2))
print("XTRA Trees Balanced Accuracy: %0.2f (+/- %0.2f)" % (np.mean(scores['test_balanced_accuracy']), np.std(scores['test_balanced_accuracy']) * 2))
print("XTRA Trees Percision: %0.2f (+/- %0.2f)" % (np.mean(scores['test_precision_macro']), np.std(scores['test_precision_macro']) * 2))
print("XTRA Trees Recall: %0.2f (+/- %0.2f)" % (np.mean(scores['test_recall_macro']), np.std(scores['test_recall_macro']) * 2))
print("XTRA Trees F1: %0.2f (+/- %0.2f)" % (np.mean(scores['test_f1_macro']), np.std(scores['test_f1_macro']) * 2))
```

```python
scoring = ['precision_macro', 'recall_macro', 'f1_macro', 'balanced_accuracy', 'accuracy']
scores = cross_validate(estimator=adaboost,
                         X=X_cs,
                         y=y_cs,
                         cv=folds,
                         scoring=scoring)
print("AdaBoost   Accuracy: %0.2f (+/- %0.2f)" % (np.mean(scores['test_accuracy']), np.std(scores['test_accuracy']) * 2))
print("AdaBoost   Balanced Accuracy: %0.2f (+/- %0.2f)" % (np.mean(scores['test_balanced_accuracy']), np.std(scores['test_balanced_accuracy']) * 2))
print("AdaBoost   Percision: %0.2f (+/- %0.2f)" % (np.mean(scores['test_precision_macro']), np.std(scores['test_precision_macro']) * 2))
print("AdaBoost   Recall: %0.2f (+/- %0.2f)" % (np.mean(scores['test_recall_macro']), np.std(scores['test_recall_macro']) * 2))
print("AdaBoost   F1: %0.2f (+/- %0.2f)" % (np.mean(scores['test_f1_macro']), np.std(scores['test_f1_macro']) * 2))
```

```python
scoring = ['precision_macro', 'recall_macro', 'f1_macro', 'balanced_accuracy', 'accuracy']
scores = cross_validate(estimator=xgboost,
                         X=X_cs,
                         y=y_cs,
                         cv=folds,
                         scoring=scoring)
print("XGBoost    Accuracy: %0.2f (+/- %0.2f)" % (np.mean(scores['test_accuracy']), np.std(scores['test_accuracy']) * 2))
print("XGBoost    Balanced Accuracy: %0.2f (+/- %0.2f)" % (np.mean(scores['test_balanced_accuracy']), np.std(scores['test_balanced_accuracy']) * 2))
print("XGBoost    Percision: %0.2f (+/- %0.2f)" % (np.mean(scores['test_precision_macro']), np.std(scores['test_precision_macro']) * 2))
print("XGBoost    Recall: %0.2f (+/- %0.2f)" % (np.mean(scores['test_recall_macro']), np.std(scores['test_recall_macro']) * 2))
print("XGBoost    F1: %0.2f (+/- %0.2f)" % (np.mean(scores['test_f1_macro']), np.std(scores['test_f1_macro']) * 2))
```
